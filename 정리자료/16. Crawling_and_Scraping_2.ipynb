```python
files = []
for i in range(10000):
    files.append(open('boa.txt').close()) #OS Error가 뜬다. -> 한꺼번에 열수 있는 파일 개수가 정해져 있다.
#파일을 열면 닫아줘야하는데 python에서는 garbage collector가 자동으로 청소해준다.
```

```python
with open('boa.txt', encoding='utf-8') as f: #with는 scope를 만들어준다. with구문에서 file을 오픈해주면 with내에서만 파일 열림.
          print(f.read())
#with구문을 쓰면 contextmanager가 알아서 file을 오픈하고 scope밖에 나가면 close해준다. -> try-finally구문도 똑같은 역할을 해준다.
```
