pandas 기본 포맷이 바(-)이기 때문에 바로 나타낼 것이다.

```python
s = pd.Series([10,100,1000,10000])
pd.to_datetime(s, unit='D') 
# unit = 'D' -> Series객체 내의 1970-01-01부터 날짜로 인식 
```

```python
s = pd.Series(['12-5-2015', '14-1-2013', '20/12/2017', '40/23/2017'])
pd.to_datetime(s, dayfirst=True, errors = 'coerce')
# dayfirst=True -> 위에 Series객체의 date를 날짜부터 인식한다.
# errors=coerce -> 잘못 파싱 되는 것은 NaT 값을 주게 된다.
```

```python
pd.Timedelta(days=5, minutes=7.34) # 5일과 7.34분 간의 시간 Timedelta 객체
```

```python
pd.Timedelta('5 days', errors='coerce') # 5일 간의 시간 Timedelta 객체
```

```python
pd.to_timedelta('67:14:45.454') 
# 67hours -> 2 days 19 hours
# 14min -> 14 min
# 45.453sec -> 45.454000 sec
```

```python
s = pd.Series([10,100])
pd.to_timedelta(s, unit='s') # unit = s -> Series객체의 숫자를 초로 인식한다
#아래의 Series 객체를 보면 10초, 1분 40초 (100초) 로 데이터가 나타났음을 알 수 있다.
```

```python
time_strings = ['2 days 24 minutes 89.67 seconds', '00:45:23.6']
pd.to_timedelta(time_strings)
# timedelta형으로 만들어진 Index객체가 만들어짐.
```

datetime으로 객체를 바꾸는 이유는 어떠한 추가기능을 쓰기 위해서..
요일을 뽑아내서 요일별 범죄율 같은 것들을 뽑아낼 수 있다

```python
crime = crime.set_index('REPORTED_DATE')
# 'REPORTED_DATE' Column을 Index로 세팅해주었다.
crime.head()
```

```python
crime.sort_index(inplace=True)
# Index를 정렬해줌.
```

```python
crime
```

```python
crime['2012-01-01':'2013-01-01']['IS_CRIME'].value_counts()[0]
# 2012-01-01부터 2013-01-01까지의 'IS_CRIME' Column에서 0값을 가지는 value 개수
```

```python
crime.loc['2016-05-12'].shape
# shape가 (243, 7) -> 243개의 rows, 7개의 columns
```

```python
crime.loc['2015-12'].sort_index()
# 2015-12월에 해당하는 자료를 sorting해준 상태.
```

```python
crime.between_time('2:00','5:00', include_end = False).head()
# 2시부터 5시 사이 시간의 data를 갖고옴.
```

```python
crime.at_time('2:00') # 딱 2시의 데이터를 갖고 온다.
```

```python
crime.first('5D')
# 제일 빠른 날짜부터 5일까지의 날짜를 포함한 데이터를 갖고옴.
```

```python
crime.resample('W') # 주별로 통계를 내준다. 일별로 -> D  월별로 -> M
```
