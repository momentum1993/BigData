agent,header 및 네트워크에 대한 정보는 F12키로 들어가서 Network 탭에서 확인할 수 있다.

```python
from urllib.request import Request
from urllib.error import HTTPError

#api가 아닌 일반적으로 사이트에 접근하는 방법
def download(url, agent="python bot", num_retries=2):
    headers = {"User-agent": agent}
    req = Request(url, headers=headers)    
    try:
        resp = urlopen(req)
    except HTTPError as e:
        resp = None
        print(e.code, e.reason, e.headers) #에러의 코드, 이유, 헤더 출력 -> agent에 대한 내용이 없다. 어떤식으로 접근 했는지에 대한 내용이 없음
        
        if 500<=e.code<600 and num_retries>0:
            return download(url, num_retries=num_retries-1)
        
    return resp

html = download("http://httpstat.us/500") #header 정보가 안들어가서 에러뜸
```

```python
import requests

def download(url, agent="python bot", num_retries=2):
    headers = {"User-agent": agent} 
    #get 방식은 바이트로 바꿔서 보낸다. 많은 데이터를 보낼때 한계가 있음.
    #post방식은 많은 데이터를 보낼 때 이용한다. 데이터를 직렬화해서 쓴다.
    resp = requests.request("get",url, headers=headers) 
    #url 오픈 할 필요가없다 위에 것에 비해 훨씬더 abstract 된 상위버전 ,코드가 더 간결해짐.
  
    if 500<=resp.status_code<600 and num_retries>0:
        print(resp.status_code, resp.reason)
        return download(url, num_retries=num_retries-1)
    
    return resp
```

```python
import requests

def downloadJson(url, agent="python bot", num_retries=2):
    headers = {"User-agent": agent} 
    #get 방식은 바이트로 바꿔서 보낸다. 많은 데이터를 보낼때 한계가 있음.
    #post방식은 많은 데이터를 보낼 때 이용한다. 데이터를 직렬화해서 쓴다.
    resp = requests.request("post",url, headers=headers, json=jsonParams) 
    #url 오픈 할 필요가없다 위에 것에 비해 훨씬더 abstract 된 상위버전 ,코드가 더 간곌해짐.
  
    if 500<=resp.status_code<600 and num_retries>0:
        print(resp.status_code, resp.reason)
        return download(url, num_retries=num_retries-1)
    
    return resp
```

```python
import requests

def downloadGet(url, agent="python bot", num_retries=2):
    headers = {"User-agent": agent} 
    #get 방식은 바이트로 바꿔서 보낸다. 많은 데이터를 보낼때 한계가 있음.
    #post방식은 많은 데이터를 보낼 때 이용한다. 데이터를 직렬화해서 쓴다.
    resp = requests.get(url, headers=headers, params=params) 
    #url 오픈 할 필요가없다 위에 것에 비해 훨씬더 abstract 된 상위버전 ,코드가 더 간결해짐.
  
    if 500<=resp.status_code<600 and num_retries>0:
        print(resp.status_code, resp.reason)
        return download(url, num_retries=num_retries-1)
    
    return resp
```
