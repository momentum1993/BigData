```python
%%writefile test.html

<!doctype html> #이것은 HTML5 규약을 따르는 문서이다.
<html lang="ko"> #이것은 "Korea" 언어팩으로 써서 읽어줘야한다고 명시하는것.사용자를 위함
    <head>
        <meta charset="UTF-8"/> #문서는 UTF-8로 인코딩 되었다.
        <title>예제</title> #crawler가 읽을 때 제일 먼저 접근. 페이지의 title
        <meta name="viewport" content="width=device-width, initial-scale=1.0"/>
        #현재 너비를 기기의 너비에 맞춰줌. 메타 태그를 여기서 안 닫아줄 경우, 인코딩이 제대로 안 되는 경우가 있음.
        <meta name="keywords" content="실습, 데이터진흥원, 노잼"/>
        #어떠한 유저가 content 내의 키워드로 검색시에 잘 뜰 수 있도록 해주는 것임.
        <script></script> #script 언어(JavaScript)
        <style>
            * {margin: 0; padding: 0;}
        </style> # style
    </head>
    <body>
     #header, section, aside, footer는 모두 블락 요소이다.
        #이렇게 태그만 보고서도 의미론적 관점에서 이해할 수 있다.(2단 구조)
        
        <ul> #단지 목록을 구성하는 wrapper. 순서 없는 리스트
            <li><span style="color:red;">첫</span>번<div style="color:blue">째</div></li> # 첫번째 리스트
            #<span>의 경우 inline 요소이기 때문에 구역(블락)을 지니지 않는다.
            #다만 <div>의 경우 블락을 지니기 때문에 한칸 띄워서 나온다.
            <li>두번째</li>
            <li>세번째</li>
        </ul> 
        #기본 속성이 Block이기 때문에 구역을 지정한다. <p> <article> 등은 블락(구역)을 지닌다. positioning하는데 편리..
        <ol> #단지 목록을 구성하는 wrapper. 순서 있는 리스트
            <li>첫번째</li> # 첫번째 리스트
            <li>두번째</li>
            <li>세번째</li>
        </ol>
        <p>
            <img src="image1.jpg", usemap="#test", style="width:100%;"/> 
            #이미지를 갖고옴. "test" 네임스페이스의 map을 갖고옴. 너비를 기기의 화면에 딱 맞게 해놓음.
        </p>
        <map name="test">
            <area shape="rect", coords="0,0,200,200", alt="오승환",href="https://www.naver.com">
            #링크를 사각형으로 (0,0), (100,100)을 눌렀을 때 해당 경로로 보내주고, alt에는 도움말을 준다.
        </map>
        <h1>첫번째 예제</h1> # Heading
        <h2>첫번째 예제</h2>
        <h3>첫번째 예제</h3> 
        <h4>첫번째 예제</h4>
        <h5>첫번째 예제</h5>
        <h6>첫번째 예제</h6> #중요도에 따라 태그를 정하고 나중에 폰트 크기를 지정하거나 할것.
        <section>
            <article>
                <h1>제목</h1>
                <p>내용</p>
            </article>
            <article>
                <h1>제목</h1>
                <p>내용   
                      ㅇㅇㅇ</p> #내용과 ㅇㅇㅇ 사이의 white space는 다 무시되어 출력됨
            </article>
            <article>
                <h1>제목</h1>
                <pre>내용   
                      ㅇㅇㅇ</pre> #<pre>태그를 통해 whitespace 포함해서 출력하게 해줌.
            </article>
            <article>
                <h1>제목</h1>
                <p>
                    단락<br/><br/>="\n\n"나누고싶다 #단락을 나눠줌.
                </p> 
            </article>
             <article>
                <h1>제목</h1>
                <p>
                    <a href="https://naver.com" target="_self">단락</a><br/><br/>="\n\n"나누고싶다 #단락을 나눠줌.
                    #<a> 라는 anchor tag를 통해 네이버로 링크를 줌. 
                    #target은 "_self"일때 자신의 페이지에서 링크 이동. "blank"일때 새창으로 띄워줌
                </p> 
            </article>
        </section>
    </body>
</html>
```

```python
%%writefile css.html
<!doctype html>
<html>
    <head>
        <style>
            div {background-color:rgba(255, 0, 0, .5);} 
            #id {color:rgb(0,0,255);}  
            .class {background-color: yellow;}
            div + p.class {background-color: green; }
            p.class {font-family:"맑은고딕", "Malgun Gothic", BatangChe, Sans-serif;}
            p.class {font-size:3.0em;}
            p.class {font-weight:bold;}
        </style>
        #div에 대해서 블록을 빨간색으로 바꿈
        #id값을 받는 것의 글씨가 파란색으로 바뀜
        #class에 대해서 블록을 노란색으로 바뀜.
        #div의 동등선상에 p.class가 뒤에 있는 것은 초록색으로 바뀐다.
        #일반적으로 프로그램은 위에서부터 코드를 읽지만 css는 딱히 그렇진 않음.
        #그래서 더욱 detail하게 선언해주면 상세하게 적용된 것이 적용 될 것임. ex) p.class
        #font가 없을 경우를 대비해서 뒤로 여러개를 나열해줌..
        #폰트를 배율로서 크기 지정해줌 (em), 굵게 해준다 (bold)
    </head>
    <body>
        <div id = "id">Content1</div>
        <div class = "class">content2</div>
        <p class = "class">content3</p>
    </body>
</html>
```
