왜도??? 첨도????   
  
<b>왜도</b> : 분포의 비대칭 정도를 나타내는 통계량이다. 정규분포와 같은 대칭 분포의 경우 왜도는 0이된다.  
<b>첨도</b> : 첨도는 분포의 꼬리부분 길이와 중앙 부분의 뾰족함에 대한 정보를 제공하는 통계량이다. 정규분포의 첨도는 0이다.
