```python
import json
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
import re

with open("account.json", "r", encoding="utf-8") as f:
    account = json.load(f)

driver = webdriver.Chrome("D:/chromedriver")
driver.get("https://sports.news.naver.com/kbaseball/index.nhn")
driver.implicitly_wait(1)

#driver.find_element_by_css_selector("ul.home_news_list li a span").click()
driver.find_element_by_css_selector("#_bottom_column_box_0 div a span.mask").click()
temp = driver.find_element_by_id("newsEndContents")
text = temp.text

text = re.sub("[^ㄱ-ㅎㅏ-ㅣ가-힣0-9A-Za-z ]"," ", text) #한글 제외한 글자들은 ""로 바꿔줌
#특수문자 없애버림

text
```

```python
from konlpy.tag import Twitter
from matplotlib import pyplot as plt
from matplotlib import font_manager, rc
import nltk
t = Twitter()

font = font_manager.FontProperties(fname = "C:/Windows/Fonts/malgun.ttf").get_name()
rc("font", family=font)

wordList = t.morphs(text)
ko = nltk.Text(wordList)
```

```python
ko.vocab()
```

```python
ko.plot(10)
```

```python
ko.dispersion_plot(["야구", "감독", "코치"])
```

```python
ko.concordance("이글스", lines=10) #'이글스'라는 단어가 들어간 10개의 라인 출력해준다.
```
